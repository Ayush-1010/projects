## For Marksheet 1

```
Image.open('/content/ms-1.jpg')
```

```
def marks(image_path):
  image = image_path
  final_extract = pytesseract.image_to_string(Image.open(image))
  sep_final = final_extract.split(sep='\n')
  sep_final2 = [ele for ele in sep_final if ele.strip()]
  l = []
  for i in sep_final2[-5:]:
    m = i.split(sep=',')[0]
    l.append(m)
  m = []
  for i in l:
    k = w2n.word_to_num(i)
    m.append(k)
  print(m)

```

```
# Running the function

marks('/content/ms-1.jpg')
```

## For Marksheet 4

```
image2 = '/content/ms-4.jpg'
```

```
Image.open(image2)
```

```
def marks2(image_path):
  image = image_path

  final_extract = pytesseract.image_to_string(Image.open(image))

  sep_final = final_extract.split(sep='\n')

  sep_final2 = [ele for ele in sep_final if ele.strip()]

  sep_final3 = sep_final2[-7:-5]

  t=[]
  for i in sep_final2[-7:-4]:
    if str(':') in list(i):
      m = i.split(sep=':')[1]
      t.append(m)
    else:
      y= i[9:]
      t.append(y)

    k = []
    for i in t:
      w = w2n.word_to_num(i)
      k.append(w)

    total = 0
    for ele in range(0, len(k)): 
      total = total + k[ele]

      if round((total/300)*100) >= 90:
        print(round((total/300)*100),'%')
        print('Campus - Alpha')
      elif  (round((total/300)*100) >= 80) and (round((total/300)*100) < 90):
        print(round((total/300)*100),'%')
        print('Campus - Beta')
      elif  (round((total/300)*100) >= 70) and (round((total/300)*100) < 80):
        print(round((total/300)*100),'%')
        print('Campus - Gamma')  
         
```
